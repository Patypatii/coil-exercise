# The sales dataset

Sales.csv can be found in the files folder. The file contains 100 sales records. Each record contains 6 fields:
* Quantity
* InvoiceDate
* UnitPrice
* CustomerID
* CountryCode

**Tasks**  
Analyze the dataset using only NumPy. 

**Step 1**   
Load the dataset using np.loadtxt() or np.genfromtxt().

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Load the dataset
raw_data = np.genfromtxt('files/sales.csv', delimiter=',', dtype=object, encoding=None, skip_header=1)
quantity = raw_data[:, 0].astype(float).astype(int)
unit_price = raw_data[:, 2].astype(float)
customer_id = raw_data[:, 3].astype(float)
country_code = raw_data[:, 4].astype(object)
invoice_date = raw_data[:, 1].astype('datetime64[D]')
```

**Step 2**   
Calculate basic statistics for the field UnitPrice:   
These basic statistics include min, max, mean, median and standard deviation.

```python
print(f'Min UnitPrice: {np.min(unit_price)}')
print(f'Max UnitPrice: {np.max(unit_price)}')
print(f'Mean UnitPrice: {np.mean(unit_price)}')
print(f'Median UnitPrice: {np.median(unit_price)}')
print(f'Std UnitPrice: {np.std(unit_price)}')
```

**Step 3**   
Revenue analysis:
* Calculate the overall total revenue
* Calculate the total revenue per country

```python
revenue = quantity * unit_price
total_revenue = np.sum(revenue)
print(f'Total overall revenue: {total_revenue}')

countries = np.unique(country_code)
for country in countries:
    mask = country_code == country
    print(f'Total revenue for {country}: {np.sum(revenue[mask])}')

# Graph: Revenue per country
plt.figure(figsize=(10, 6))
country_revenue = [np.sum(revenue[country_code == c]) for c in countries]
plt.bar(list(map(str, countries)), country_revenue)
plt.title('Revenue per Country')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

**Step 4**   
Customer insights:
* Identify the customer with the highest total spend.
* Identify the customer with the most purchases.

```python
unique_customers = np.unique(customer_id)
customer_spend = {cid: np.sum(revenue[customer_id == cid]) for cid in unique_customers}
highest_spender = max(customer_spend, key=customer_spend.get)
print(f'Highest Spender (ID): {highest_spender} with {customer_spend[highest_spender]}')

customer_purchases = {cid: np.sum(customer_id == cid) for cid in unique_customers}
most_purchases = max(customer_purchases, key=customer_purchases.get)
print(f'Most Purchases (ID): {most_purchases} with {customer_purchases[most_purchases]} transactions')
```

**Step 5**   
Time analysis:
* Convert timestamps to NumPy datetime64.
* Group by month and calculate monthly revenue.

```python
# Convert to monthly
months = invoice_date.astype('datetime64[M]')
unique_months = np.unique(months)
monthly_revenue = [np.sum(revenue[months == m]) for m in unique_months]

for m, r in zip(unique_months, monthly_revenue):
    print(f'Revenue for {m}: {r}')

# Graph: Monthly Revenue
plt.figure(figsize=(10, 6))
plt.plot(unique_months.astype(str).tolist(), monthly_revenue, marker='o')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```
